```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("Seasonality.csv")
data.head(5)
```

```python
# Ensure the date column is in datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Set the date column as the index
data.set_index('Date', inplace=True)

# Group by date and calculate the mean for each group
daily_data = data.groupby(data.index).mean()

# Plot the time series for NDVI, SAVI, and EVI with increased font size
plt.figure(figsize=[15, 7])

plt.plot(daily_data.index, daily_data['NDVI'], label='NDVI', marker='o')
plt.plot(daily_data.index, daily_data['SAVI'], label='SAVI', marker='o')
plt.plot(daily_data.index, daily_data['EVI'], label='EVI', marker='o')

plt.xlabel('Date', fontsize=15)
plt.ylabel('Index Value', fontsize=15)
#plt.title('Time Series Comparison of NDVI, SAVI, and EVI', fontsize=16)
plt.legend(fontsize=15)
plt.grid(True)

# Increase font size of tick labels
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.tight_layout()

# Save the plot to a file
#output_path = 'path/to/save/daily_ndvi_savi_evi_time_series_with_font.png'  # Update with your desired output path
#plt.savefig(output_path)

# Display the plot
plt.show()
```
